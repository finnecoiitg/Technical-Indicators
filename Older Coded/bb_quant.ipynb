```python
df['upperBB'], df['middleBB'], df['lowerBB'] = talib.BBANDS(df['Adj Close'], nbdevup=2, nbdevdn=2, timeperiod=20, matype = 0) 

#the matype = 0 - means SMA, u can read abt varying the number
#i couldnt find anything as such lol

df
```

```python

#at this pt for aesthetic purposes u can do df.dropna()

df = df.dropna()
df
```

```python
df['SMA'] = df['Adj Close'].rolling(20).mean()
df['stddev'] = df['Adj Close'].rolling(20).std()
df['lower_band'] = df['SMA'] - 2*df['stddev']
df['upper_band'] = df['SMA'] + 2* df['stddev']
df

#basically coded the thing with basic math and df
```

```python
df
```

```python
df.plot.line(x ='date',y = ['Adj Close', 'SMA','upperBB','lowerBB'])
```
